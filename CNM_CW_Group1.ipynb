```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np

x = np.linspace(0, 10, 10) #position in river
conc = np.ones([10, 10]) * x #polutant concentration at point
data = pd.DataFrame(conc)

interval = 0.01

fig, ax = plt.subplots(figsize=(4,3))


plt.rcParams["animation.html"] = "jshtml"
plt.rcParams['figure.dpi'] = 150
plt.ioff() # interactive off


def animate(t):
  plt.cla()
  y = np.array(data[t])
  plt.plot(x, y)
  plt.xlim(0,10)
  plt.ylim(0, 10)
  ax.set_title("Concentration of pollutant at t = " + str(interval*t) + "s", fontsize=10, verticalalignment='top')
  plt.xlabel("x", fontsize=8)
  plt.ylabel("Concentration", fontsize=8)

matplotlib.animation.FuncAnimation(fig, animate, frames=10)

```
